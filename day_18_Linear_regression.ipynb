```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("/content/Salary_dataset.csv")
df.head()
```

```python
df_new=df.drop(columns=["Unnamed: 0"])
df_new
```

```python
x=df_new.iloc[:,:-1].values  # iloc index location  duvide
y=df_new.iloc[:,1].values # value convert dataframe to array
```

```python
from sklearn.model_selection import train_test_split   # import
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=.2)
```

```python
x_test
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train) # fit is give data like as
```

```python
lr.score(x_test,y_test) # score give the accuracy
```

```python
y_pred=lr.predict(x_test) # do the prediction
x_pred=lr.predict(x_train)
```

```python
lr.score(x_test,y_test)
```

```python
plt.scatter(x_train,y_train,color="red")
plt.plot(x_train,x_pred,color="green")
plt.title("salary vs experience")
plt.xlabel("years of experience")
plt.ylabel("salary")
plt.show()
```
