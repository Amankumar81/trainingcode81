```python
x=np.array(["A","B","C","D","E"])
y=np.array([3,6,8,9,2])
plt.bar(x,y)
plt.show()
```

```python
x=np.array(["A","B","C","D","E"])
y=np.array([3,6,8,9,2])
plt.bar(x,y,color="red")
plt.show()
```

```python
y=np.array([35,25,25,15])
name=["apple","mango","lichi","melon"]
plt.pie(y,labels=name,) # name give
plt.show()
```

```python
y=np.array([35,25,25,15])
name=["apple","mango","lichi","melon"]
myexplode=[0.2,0,0,0]
plt.pie(y,labels=name,explode=myexplode)  # take a difference between a element
plt.show()
```

```python

y=np.array([35,25,25,15])
name=["apple","mango","lichi","melon"]
myexplode=[0.2,0,0,0]
plt.pie(y,labels=name,explode=myexplode,shadow=True)
plt.show()
```

```python
y=np.array([35,25,25,15])
name=["apple","mango","lichi","melon"]
myexplode=[0.2,0,0,0]
mycolour=["blue","red","pink","black"]
plt.pie(y,labels=name,colors=mycolour)  # take a differenct colour
plt.show()
```

```python
y=np.array([35,25,25,15])
name=["apple","mango","lichi","melon"]
myexplode=[0.2,0,0,0]
plt.pie(y,labels=name,explode=myexplode)  # take a difference between a element
plt.legend()  # take legend
plt.show()
```

```python
y=np.array([35,25,25,15])
name=["apple","mango","lichi","melon"]
myexplode=[0.2,0,0,0]
plt.pie(y,labels=name,explode=myexplode)  # take a 
plt.legend(title="four furits:")
plt.show()
```
