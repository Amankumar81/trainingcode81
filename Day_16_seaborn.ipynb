### dealing eith outlier
# 1. remove

```python
upper_limit=tip['total_bill'].quantile(0.9)# quantile range 0 t0 1 and default value is 0.5 and us float and array value
upper_limit
```

```python
sns.boxplot(tip[tip['total_bill']<upper_limit]['total_bill'])  # clean the outlier to make easy
```

```python
### usnig IQR range
q1=np.percentile(tip['total_bill'],25) # iqr range is 25 to 75 and 25 mean q1 and 75 mean q3 and percentile is function of np
q3=np.percentile(tip['total_bill'],75)
IQR=q3-q1

upper_limit=q3+1.5*IQR
lower_limit=q1-1.5*IQR
```
